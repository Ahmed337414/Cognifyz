```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('C:\\Users\\ahmed\\Downloads\\Cognifyz-Internship\\Cognifyz-DS\\Level-3_Tasks\\Dataset.csv')
```

We ensured 'Aggregate rating' was numeric and dropped rows with missing ratings. Then, we visualized the distribution of ratings through a histogram, providing insights into the frequency distribution of restaurant ratings.

```python
# Convert 'Aggregate rating' to numeric, in case it is not
data['Aggregate rating'] = pd.to_numeric(data['Aggregate rating'], errors='coerce')

# Drop rows with missing ratings
data = data.dropna(subset=['Aggregate rating'])

# Histogram of Ratings
plt.figure(figsize=(10, 6))
plt.hist(data['Aggregate rating'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot of Ratings
plt.figure(figsize=(10, 6))
sns.boxplot(data['Aggregate rating'])
plt.title('Box Plot of Ratings')
plt.xlabel('Rating')
plt.show()
```

We processed and extracted relevant data, focusing on ratings and cuisines. By calculating the average rating for each cuisine, we visualized the top 20 cuisines with the highest average ratings in a bar plot, providing insights into culinary preferences.

```python
# Preprocess and extract relevant data for ratings and cuisines
cuisine_ratings = data[['Cuisines', 'Aggregate rating']].dropna()
cuisine_ratings['Cuisines'] = cuisine_ratings['Cuisines'].str.split(', ')
cuisine_ratings = cuisine_ratings.explode('Cuisines')

# Group by cuisine and calculate the average rating
cuisine_avg_ratings = cuisine_ratings.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

# Bar Plot of Average Ratings by Cuisine
plt.figure(figsize=(14, 8))
cuisine_avg_ratings.head(20).plot(kind='bar', color='skyblue')
plt.title('Average Ratings by Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

We grouped the data by city and computed the average rating for each city. The bar plot displays the top 20 cities with the highest average ratings, offering insights into regional variations in restaurant quality.

```python
# Group by city and calculate the average rating
city_avg_ratings = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

# Bar Plot of Average Ratings by City
plt.figure(figsize=(14, 8))
city_avg_ratings.head(20).plot(kind='bar', color='coral')
plt.title('Average Ratings by City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter Plot of Price Range vs. Rating
plt.figure(figsize=(10, 6))
plt.scatter(data['Price range'], data['Aggregate rating'], alpha=0.6, edgecolors='w', linewidth=0.5)
plt.title('Price Range vs. Rating')
plt.xlabel('Price Range')
plt.ylabel('Rating')
plt.show()
```

```python
# Box Plot of Rating Distribution by Has Table Booking
plt.figure(figsize=(10, 6))
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=data)
plt.title('Rating Distribution by Has Table Booking')
plt.xlabel('Has Table Booking')
plt.ylabel('Rating')
plt.show()
```
