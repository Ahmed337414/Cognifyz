# Task: Table Booking and Online Delivery

1. Determine the percentage of restaurants that offer table booking and online delivery.
2. Compare the average ratings of restaurants with table booking and those without.
3. Analyze the availability of online delivery among restaurants with different price ranges.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('Dataset .csv')
data
```

In this part of our analysis, we focus on the availability of table booking and online delivery services among the restaurants in our dataset.

First, we calculate the percentage of restaurants that offer table booking. We do this by finding the proportion of restaurants that have 'Yes' for table booking and converting it into a percentage.

Next, we similarly determine the percentage of restaurants that provide online delivery services. We calculate the proportion of restaurants with 'Yes' for online delivery and convert it into a percentage.

Finally, we print these percentages, formatted to two decimal places, to clearly present the data.

```python
# Calculate the percentage of restaurants that offer table booking
table_booking_percentage = (data['Has Table booking'].value_counts(normalize=True) * 100).loc['Yes']

# Calculate the percentage of restaurants that offer online delivery
online_delivery_percentage = (data['Has Online delivery'].value_counts(normalize=True) * 100).loc['Yes']

print(f"Percentage of restaurants that offer table booking: {table_booking_percentage:.2f}%")
print(f"Percentage of restaurants that offer online delivery: {online_delivery_percentage:.2f}%")
```

The average ratings of restaurants that offer table booking and those that do not by filtering the dataset based on the "Has Table booking" column and calculating the mean of the "Aggregate rating" for each group. The results are then printed, showing the average rating for restaurants with and without table booking.

```python
# Compare the average ratings of restaurants with and without table booking
avg_rating_with_table_booking = data[data['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
avg_rating_without_table_booking = data[data['Has Table booking'] == 'No']['Aggregate rating'].mean()

print(f"Average rating of restaurants with table booking: {avg_rating_with_table_booking:.2f}")
print(f"Average rating of restaurants without table booking: {avg_rating_without_table_booking:.2f}")
```

The ailability of online delivery among restaurants with different price ranges by grouping the data by price range and calculating the percentage of restaurants that offer online delivery within each range. The results are then plotted as a bar chart to visualize the availability of online delivery by price range.

```python
# Analyze the availability of online delivery among restaurants with different price ranges
online_delivery_by_price_range = data.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0) * 100

print(online_delivery_by_price_range)

# Plot the results
plt.figure(figsize=(10, 6))
online_delivery_by_price_range.plot(kind='bar', stacked=True, color=['orange', 'skyblue'])
plt.title('Availability of Online Delivery by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Has Online delivery', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
