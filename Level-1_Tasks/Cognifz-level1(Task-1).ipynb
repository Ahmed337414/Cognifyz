## Analyzing the Target Variable

```python
# Fill missing values in the "Cuisines" column with "Unknown"
data['Cuisines'].fillna('Unknown', inplace=True)

# Check data types of each column
print(data.dtypes)

# Analyze the distribution of the target variable "Aggregate rating"
aggregate_rating = data['Aggregate rating'].value_counts().sort_index()

# Display the data types and the aggregate_rating of the target variable
print(aggregate_rating)
```

```python
import matplotlib.pyplot as plt

# Plot the "Aggregate rating"
plt.figure(figsize=(10, 8))
aggregate_rating.plot(kind='bar', color='skyblue')
plt.title('Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
